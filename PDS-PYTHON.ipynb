***1 Señales y sistemas***
> 1.1 Definición y tipos de señales  
Podemos definir a una señal como aquella que tiene una magnitud (escarlar vectorial) que varía en función de otra magnitud o magnitudes como puede ser el tiempo, la distancia, etc.
Éstas se pueden clasificar en señales determinísticas y aleatorias.
Como ejemplo de señales podemos mencionar las acústicas, el electrocardiograma, ondas sísticas, etc.
Las señales del mundo natural son continuas en el tiempo y las señales discretas son valores secuenciales en cada instante de tiempo. Tener en cuenta que las señales discretas tienen valor indeterminado entre cada muestra, no confundir con ponerle valor 0. Por ejemplo la temperatura del ambiente es una señal continua, y el valor de depreciación de una casa a lo largo del tiempo es discreta, porque sólos se contemplan valores por cada año.

##### *Señal paso unitario = Unit step*   
$
u_{s}(t) = \left\{
    \begin{array}\\
        1 & \mbox{for } \ t \geq 0 \\
        0 & \mbox{for } \ t <0
    \end{array}
\right.
$

```python
t = np.arange(-12,12, 0.5)
ceros = np.zeros(int(len(t)/2))
ones = np.ones(int(len(t)/2))
step = []
step = np.append(ceros, ones)
plt.figure(figsize=(10,4))
plt.plot(t, step, t, step, 'ro')
plt.legend(['Señal continua', 'Señal discreta'])
plt.title('Señal escalón unitario')
plt.grid ('on')
```

```python
plt.figure(figsize=(15,20))
plt.subplot(2,1,1)
plt.plot(t,step)
plt.title('Señal continua')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid('on')
plt.subplot(2,1,2)
plt.stem(t,step, use_line_collection=True)
plt.title('Señal discreta')
plt.xlabel('Secuencia')
plt.ylabel('Amplitud')
plt.grid('on')
```
